```python
import pandas as pd
```

```python
arena = pd.read_table('./Resources/arena.txt')
```

```python
arena_df = pd.DataFrame(arena)
```

```python
arena_df = arena_df.rename(columns={arena_df.columns[0]: "Data" })
arena_df = arena_df[~arena_df['Data'].str.contains('Location|Phone|Website|Email|Founded|Arena:')]
arena_df = arena_df.replace('- ', '-2019', regex=True)
arena_df = arena_df.replace(', ', ',', regex=True)
arena_df = arena_df[arena_df['Data'].str.contains('-')]
arena_df[['Year','Venue']] = arena_df["Data"].str.split(" ", 1, expand=True)
#arena_df = arena_df[['Year','Venue']]
arena_df['Capacity'] = arena_df['Venue'].str.split().str[-1]
arena_df['City'] = arena_df['Venue'].str.split().str[-2]
#arena_df = arena_df[['Year','Venue','City','Capacity']]
arena_df['Venue'] = [e.replace(k, '') for e, k in zip(arena_df.Venue.astype('str'), arena_df.Capacity.astype('str'))]
arena_df['Venue'] = [e.replace(k, '') for e, k in zip(arena_df.Venue.astype('str'), arena_df.City.astype('str'))]
arena_df[['StartYear','EndYear']] = arena_df["Year"].str.split("-", 1, expand=True)
arena_df = arena_df[['StartYear','EndYear','Venue','City','Capacity']]
arena_df.head(10)
```

```python
arena_df.to_csv('Resources/Arena_list.csv')
```
