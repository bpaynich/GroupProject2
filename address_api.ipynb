```python
import pandas as pd
import json
import requests

from config import api_key
```

```python
file = ("test_csv.csv")
data = pd.read_csv(file)
#print(data)
df = pd.DataFrame(data)
df
stadium_list = df['arenas'].tolist()
stadium_name = []
stadium_state = []
stadium_lat = []
stadium_lng = []
```

```python
for stadium_search in stadium_list:
    
    #stadium_search = "name"


    params = {
        "address": stadium_search,
        "key": api_key
    }

    base_url = "https://maps.googleapis.com/maps/api/geocode/json"

    

    st_geo = requests.get(base_url, params).json()

    lat = st_geo["results"][0]["geometry"]["location"]["lat"]
    lng = st_geo["results"][0]["geometry"]["location"]["lng"]


    # Use lat, lng to use places API to find nearest dentist
    target_search = "stadium"

    params = {
        "location": f"{lat},{lng}",
        "types": target_search,
        "rankby": "distance",
        "key": api_key
    }

    base_url = f"https://maps.googleapis.com/maps/api/geocode/json?latlng={lat},{lng}&key={api_key}"


    # Run request
    st_reverse_geo = requests.get(base_url).json()

    print(st_reverse_geo)
    

    st = st_reverse_geo["results"][0]["formatted_address"]
    stadium_state.append(st)
    
    
    
    lt =  st_geo["results"][0]["geometry"]["location"]["lat"]
    stadium_lat.append(lt)
    
    ln =  st_geo["results"][0]["geometry"]["location"]["lng"]
    stadium_lng.append(ln)
```

```python
#(stadium_name, stadium_state, stadium_lat, stadium_lng)

st_list = {'arena name': stadium_name,
           'address': stadium_state,
           'latitude': stadium_lat,
           'longitude': stadium_lng}

stadium_df = pd.DataFrame(st_list, columns=[stadium_name, stadium_state, stadium_lat, stadium_lng])


#stadium_df  = stadium_df.rename(columns={stadium_df.columns[0]: "arena name", \
                                      #stadium_df.columns[1]: "address", \
                                     #stadium_df.columns[2]: "latitude", \
                                        #stadium_df.columns[3]: "longitude"})
#stadium_df["address"] = stadium_df['address', 'city'].str.split(",", n = 1, expand = True)
#stadium_df = pd.DataFrame(stadium_df.address.str.split(',',1).tolist(),
                                   #columns = ['address','city'])
```
